```python
import pandas as pd
Covid = pd.read_csv('/home/ANA522/mod02/01-10-2022.csv', sep = ',')
#Covid
```

```python
Covid.describe()
```

```python
##Q1.Which are the top 10 Countries (Country_Region) from the dataset recorded highest total deaths for 01-10-2022?

pivot = Covid.pivot_table(index = ['Country_Region'], values = ['Deaths'], aggfunc = 'sum')
top10 = pivot.nlargest(10, ['Deaths'])
print("Top 10 countries with highest deaths are:")
top10
```

```python
###Q2.Which are the top 3 Country_Region, Province_State (Combined_Key) that added new confirmed cases on 01-10-2022?

top3 = Covid.nlargest(3, ['Confirmed'])
top3[['Combined_Key', 'Confirmed']]
```
